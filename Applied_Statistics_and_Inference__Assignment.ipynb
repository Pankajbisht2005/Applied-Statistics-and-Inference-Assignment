Q1.  What are Type I and Type II errors in hypothesis testing, and how do they
impact decision-making?
- Type I and Type II Errors in Hypothesis Testing

When you run a hypothesis test, you start with:

Null hypothesis (H₀): No effect, no difference, status quo.

Alternative hypothesis (H₁): There is an effect or difference.

While making a decision (reject or fail to reject H₀), two types of mistakes can happen.

- Type I Error (False Positive)

Meaning: You reject the null hypothesis even though it is actually true.

Probability of making this error: α (significance level) — usually 0.05.

Example:

A company wants to test if a new marketing campaign increased sales.

H₀: New campaign did not increase sales.

H₁: New campaign did increase sales.

If the analyst concludes “Sales increased” even though, in reality, nothing improved

 Type I error
Impact: Company wastes money continuing a useless campaign.

- 2️ Type II Error (False Negative)

Meaning: You fail to reject the null hypothesis even though it is false.

Probability of making this error: β

Same campaign example:

If the analyst concludes “Campaign did NOT increase sales” even though in reality it did

 Type II error

Impact: Company stops a campaign that was actually helping.

 - Type I Error Impact:

You take action based on a wrong assumption.

Approving a bad product

Investing in a useless strategy

Triggering false fraud alerts

Deploying a bad model thinking it's better

- Type II Error Impact:

You miss out on an actual opportunity.

Rejecting a working feature

Missing real fraud

Keeping a bad model active

Losing revenue because you ignored a real improvement

Q2: What is the P-value in hypothesis testing, and how should it be interpreted
in the context of the null hypothesis?
- P-value: The P-value is the probability of getting your observed results  assuming the null hypothesis (H₀) is true.

How to Interpret the P-value
- 1 If P-value ≤ α (usually 0.05):

You have strong evidence against H₀.
You reject the null hypothesis.

it means:
Your result is unlikely to occur just by chance. Something real is going on.

- 2️ If P-value > α:

You do NOT reject the null hypothesis.

it means:
Your data is not strong enough to prove any effect.
You don’t have enough evidence to say H₀ is wrong.

Q3:Explain the difference between a Z-test and a T-test, including when to use
each.

- Difference Between Z-Test and T-Test
- 1 Z-Test

Use a Z-test when:

Sample size is large (n ≥ 30)

Population standard deviation (σ) is known

Data is approximately normally distributed

What it’s used for:

Testing population means

Testing proportions

Comparing large samples

- 2️ T-Test

Use a T-test when:

Sample size is small (n < 30)

Population standard deviation is unknown

You only have sample standard deviation (s)

What it’s used for:

Testing population mean using small samples

Comparing two sample means

Paired sample tests

Q4: What is a confidence interval, and how does the margin of error influence
its width and interpretation?
- Confidence Interval

A confidence interval (CI) is a range of values, calculated from sample data, that is likely to contain the true population parameter.

How Margin of Error Influences the Width:

The margin of error (ME) is half the width of the confidence interval.

CI = Sample Estimate ± Margin of Error

- a) If margin of error increases → CI becomes wider

It means:

Your estimate is less precise.

There’s more uncertainty.

The interval covers a bigger range.

- b) If margin of error decreases → CI becomes narrower

Meaning:

Your estimate is more precise.

You have stronger, tighter evidence.

Q5:D escribe the purpose and assumptions of an ANOVA test. How does it
extend hypothesis testing to more than two groups?
- Purpose of ANOVA

ANOVA is used to test whether three or more group means are significantly different from each other.

In simple language:
 Instead of comparing groups one by one (which becomes messy and increases error), ANOVA checks all groups at once.

Example:
Comparing average sales of 3 regions or testing 4 marketing strategies.

Null hypothesis (H₀): All group means are equal.
Alternative hypothesis (H₁): At least one group mean is different.
  
- How ANOVA Extends Hypothesis Testing to More Than Two Groups

If you compare 3 groups with t-tests, you would need multiple pairwise tests:

Group 1 vs Group 2

Group 1 vs Group 3

Group 2 vs Group 3

That increases Type I error.

 ANOVA solves this by testing all group means simultaneously with a single test.

It tells you whether a difference exists, not which groups differ—post-hoc tests handle that.

```python
#Q6  Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.

import numpy as np
from scipy.stats import norm

# Sample dataset (replace with your own values)
data = np.array([52, 49, 50, 48, 51, 53, 47, 50, 49, 52])

# Given information
population_mean = 50
population_std = 3
alpha = 0.05

# Step 1: Calculate sample statistics
sample_mean = np.mean(data)
n = len(data)

# Step 2: Compute Z-statistic
z_value = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Step 3: Calculate p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_value)))

# Step 4: Interpret the result
if p_value < alpha:
    decision = "Reject the null hypothesis (H₀). There is a significant difference."
else:
    decision = "Fail to reject the null hypothesis (H₀). No significant difference found."

# Output
print("Sample Mean:", sample_mean)
print("Z-value:", z_value)
print("P-value:", p_value)
print("Decision:", decision)
```

```python
#Q8: Generate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python.
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate non-normal population data (Exponential distribution)
population = np.random.exponential(scale=2, size=1000000)

# Step 2: Draw multiple samples and calculate their means
sample_size = 50
num_samples = 1000
sample_means = []

for _ in range(num_samples):
    sample = np.random.choice(population, sample_size)
    sample_means.append(np.mean(sample))

sample_means = np.array(sample_means)

# Step 3: Plot the population distribution (skewed)
plt.figure(figsize=(6,4))
plt.hist(population, bins=50)
plt.title("Original Population (Exponential, Skewed)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Step 4: Plot the distribution of sample
plt.figure(figsize=(6,4))
plt.hist(sample_means, bins=30)
plt.title("Distribution of Sample Means (CLT Approx Normal)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

```python
#Q9:  Write a Python function to calculate and visualize the confidence interval for a sample mean.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_confidence_interval(data, confidence=0.95):
    data = np.array(data)

    # Calculate sample statistics
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    n = len(data)

    # Z-score for confidence level
    z = norm.ppf((1 + confidence) / 2)

    # Margin of error
    margin_error = z * (std / np.sqrt(n))

    # Confidence interval
    lower = mean - margin_error
    upper = mean + margin_error

    print(f"Sample Mean: {mean}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower}, {upper})")

    # Visualization
    plt.figure(figsize=(7, 4))
    plt.axvline(lower, linestyle="--")
    plt.axvline(upper, linestyle="--")
    plt.axvline(mean, linestyle="-")
    plt.title(f"{int(confidence*100)}% Confidence Interval for Sample Mean")
    plt.xlabel("Value")
    plt.show()

    return (lower, upper)

# Example usage:
sample_data = [12, 15, 14, 13, 16, 18, 17, 14, 15, 16]
plot_confidence_interval(sample_data)

```

```python
#Q10: 10: Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.

import numpy as np
from scipy.stats import chisquare

# Observed frequencies (example dataset)
observed = np.array([30, 14, 34, 22])

# Expected frequencies (must sum to same total)
expected = np.array([25, 25, 25, 25])

# Chi-square test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square Statistic:", chi_stat)
print("P-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Observed distribution is significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference detected.")
"""Explaination of the outcome:

Null Hypothesis (H₀):

The observed distribution matches the expected distribution.

Alternative Hypothesis (H₁):

The observed distribution does not match the expected distribution."""
```
